# Model a Galaxy Cluster

## Notebook for generating an example galaxy cluster model. 

This notebook goes through the steps to generate model data for galaxy cluster weak lensing observables.  In particular, we define a galaxy cluster model that follows and NFW distribution and generate various profiles for the model (mass density, convergence, shear, etc.), which we plot.  Note, a full pipeline to measure a galaxy cluster weak lensing mass requires fitting the observed (or mock) data to a model. In this notebook we use the OO interface to modeling.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Imports specific to clmm 

```python
import sys
import os
sys.path.append('./support')

os.environ['CLMM_MODELING_BACKEND'] = 'ccl' # here you may choose ccl or nc (NumCosmo)

import clmm
from clmm import Cosmology
```

Make sure we know which version we're using

```python
clmm.__version__
```

Define a cosmology using astropy

```python
cosmo = Cosmology(H0 = 70.0, Omega_dm0 = 0.27 - 0.045, Omega_b0 = 0.045, Omega_k0 = 0.0)
```

Define the galaxy cluster model.  Here, we choose parameters that describe the galaxy cluster model, including the mass definition, concentration, and mass distribution.  For the mass distribution, we choose a distribution that follows an NFW profile.

```python
moo = clmm.Modeling (massdef = 'mean', delta_mdef = 200, halo_profile_model = 'nfw')

moo.set_cosmo (cosmo)
moo.set_concentration (4)
moo.set_mass (1.e15)

z_cl = 1.
z_source = 2.
```

Quick test of all modeling functionality

```python
#r3d        = np.logspace(-2, 2, 100)
r3d        = 1.
rho        = moo.eval_density (r3d, z_cl)
Sigma      = moo.eval_sigma (r3d, z_cl)
DeltaSigma = moo.eval_sigma_excess (r3d, z_cl)
Sigmac     = moo.eval_sigma_crit (z_cl, z_source)
gammat     = moo.eval_shear (r3d, z_cl, z_source)
kappa      = moo.eval_convergence (r3d, z_cl, z_source)
gt         = moo.eval_reduced_shear (r3d, z_cl, z_source)
mu         = moo.eval_magnification (r3d, z_cl, z_source)
```

Plot the predicted profiles

```python
def plot_profile(r, profile_vals, profile_label='rho'):
    plt.loglog(r, profile_vals)
    plt.xlabel('r [Mpc]', fontsize='xx-large')
    plt.ylabel(profile_label, fontsize='xx-large')
```

```python
plot_profile(r3d, rho, '$\\rho_{\\rm 3d}$')
```

```python
plot_profile(r3d, Sigma, '$\\Sigma_{\\rm 2d}$')
```

```python
plot_profile(r3d, DeltaSigma, '$\\Delta\\Sigma_{\\rm 2d}$')
```

```python
plot_profile(r3d, kappa, '$\\kappa$')
```

```python
plot_profile(r3d, gammat, '$\\gamma_t$')
```

```python
plot_profile(r3d, gt, '$g_t$')
```

```python
plot_profile(r3d, mu, '$\mu$')
```
